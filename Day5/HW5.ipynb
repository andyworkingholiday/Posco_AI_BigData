## Set 구현하기

```python
class Set:
    def __init__(self, items = []):
        self.items = []
        if len(items) != 0:
            for item in items:
                if item not in self.items:
                    self.items.append(item)
                    
    def __str__(self):
        printing = ''
        for item in self.items:
            printing += str(item)
            printing += ', '
        
        printing = printing[:len(printing)-2]
        return '{' + printing + '}'
    
    def __len__(self):
        return len(self.items)
    
    def __contains__(self,elem):
        if elem in self.items:
            return True
        else:
            return False
        
    def __le__(self, other):
        for item in self.items:
            if item not in other.items:
                return False
        return True
        
    def __ge__(self,other):
        for item in other.items:
            if item not in self.items:
                return False
        return True
    
    def __or__(self, other):
        dummy = Set()
        for item in self.items:
            dummy.add(item)
        for item in other.items:
            dummy.add(item)
        return dummy
        
    
    def __ior__(self, other):
        for item in other.items:
            if item not in self.items:
                self.items.append(item)
                
        return self
    
    def __and__(self, other):
        dummy = Set()
        for item in other.items:
            if item in self:
                dummy.add(item)
            
        return dummy
    
    def __iand__(self, other):
        for item in self.items:
            if item not in other.items:
                self.items.remove(item)
                
        return self
        
                
    def __sub__(self, other):
        dummy = Set()
        for item in self.items:
            dummy.add(item)
        for item in other.items:
            dummy.discard(item)
        return dummy
    
    def __isub__(self, other):
        for item in other.items:
            if item in self.items:
                self.items.remove(item)
                
        return self
    
    def add(self, elem):
        if elem not in self.items:
            self.items.append(elem)
            
    def discard(self, elem):
        if elem in self.items:
            self.items.remove(elem)
                     
    def clear(self):
        self.items = []
```

```python
a = Set([1, 2, 3, 4])
b = Set([1, 2, 3, 4])

print(a)
print(b)
print()

a.discard(4)
b.discard(1)
print(a)
print(b)
print()

print(len(a))
print(1 in a)
print(1 in b)
print()
print(a | b)
print(a & b)
print(a - b)
print()

print(a <= b)
print(a <= a | b)
print(a >= b)
print(a >= a & b)
print()

b.clear()
print(b)
print()

a = Set([1,2,3])
b = Set([3,4])
address_a = id(a)
a |= b
print(a)
print(address_a == id(a))
print()

a = Set([1,2,3])
b = Set([3,4])
address_a = id(a)
address_a = id(a)
a &= b
print(a)
print(address_a == id(a))
print()

a = Set([1,2,3])
b = Set([3,4])
address_a = id(a)
address_a = id(a)
a -= b
print(a)
print(address_a == id(a))
print()
```
